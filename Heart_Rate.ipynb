```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collections
heartdf = pd.read_csv('/content/heart.csv')
```

```python
heartdf.shape
```

```python
heartdf.head()
```

```python
#checking distrubution of target variable
heartdf['target'].value_counts()
#1 = BAD HEART
#2 = GOOD HEART
```

```python
#SPLITING THE DATA NAD TARGET
x = heartdf.drop(columns='target', axis=1)
y = heartdf['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#MODEL TRANING
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
#ACCURCY SCORE ON SEEN DATA X TRAIN
x_train_pred = model.predict(x_train)
trainaccur = accuracy_score(x_train_pred, y_train)
print(f"{trainaccur*100:.2f}%")
```

```python
#ACCURCY SCORE ON UNSEEN DATA X Test
x_test_pred = model.predict(x_test)
testaccur = accuracy_score(x_test_pred, y_test)
print(f"{testaccur*100:.2f}%")
```

**BUILDING PREDICTIVE SYSTEM**

```python
input_data = (41,0,1, 130, 204, 0, 0, 172, 0, 1.4, 2, 0, 2)
#chaning to numpu
numpy_data = np.asarray(input_data)
#reshape
reshape = numpy_data.reshape(1,-1)
prediction = model.predict(reshape)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```

**IMPORTING LIBRARY**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.value_counts('Class')
```

DATA IS UNVENE

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
fraud.shape
```

```python
legit.shape
```

NOW CREATE A NEW FILE AND ADD BOTH

```python
legit_new = legit.sample(n=492)
```

```python
legit_new.shape
```

```python
df_new = pd.concat([legit_new, fraud], axis = 0)
```

```python
df_new.shape
```

```python
x = df_new.drop(columns='Class', axis=1)
y = df_new['Class']
```

```python
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x_test.shape, x_train.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

**Checking Acuuracy**

```python
prec = model.predict(x_train)
acc = accuracy_score(prec, y_train)
```

```python
print(f'{"accuracy score"}: {accur * 100}%')
```

```python
preciss = model.predict(x_test)
accur1 = accuracy_score(preciss, y_test)
```

```python
print(f'{"accuracy score"}: {accur1 * 100}%')
```

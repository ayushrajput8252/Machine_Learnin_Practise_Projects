```python
#df['Gender', 'Marride', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status'] = LabelEncoder().fit
x = df.drop(columns=['Loan_ID', 'Loan_Status'])
y = df['Loan_Status']
```

```python
label_enc = LabelEncoder()
coloumn = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
for col in coloumn:
  x[col] = label_enc.fit_transform(x[col])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
#TRANING MODEL
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

```python
#predict value
predict = model.predict(x_test)
accuracy = accuracy_score(predict, y_test)
print(f'accuracy on test data : {accuracy*100:.0f}%')
```
